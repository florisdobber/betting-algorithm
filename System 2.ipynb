### Importing the libraries and functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
def add_entry(k, dictionary): #used to update the dictionaries
    """Either adds 1 to a given key:value pair or initializes a key:value pair"""
    if k in dictionary: #checks if there is already an entry
        dictionary[k] += 1
    else: #if not, it'll create a new entry
        dictionary[k] = 1

def shrink_df(df):
    df = df[['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]
    df.columns = ['Home_team', 'Away_team', 'Home_goals', 'Away_goals', 'Result']
    return df

def get_odds(df):
    df = df[['B365H', 'B365D', 'B365A', 'BWH', 'BWD', 'BWA','IWH','IWD', 'IWA', 'PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA']]
    return df

def add_ratings(df):
    #Add columns for ratings
    df['Home_rating'] = np.nan
    df['Away_rating'] = np.nan

    #Get the name of everytime
    team_names = df['Home_team'].unique()
    
    for team in team_names:
        #Index for every match the team played
        games = df.loc[(df['Home_team'] == team) 
                       | (df['Away_team'] == team)].index
        #Array to hold the goal difference of each match
        goal_diffs = np.empty(38, dtype=int)
        #Array to hold team rating for matches 7-38 inclusive
        team_ratings = np.empty(32, dtype=int) 
        
        #Calculate the goal difference for each match
        for index in range(len(games)):
            game = df.iloc[games[index]]
            if (game['Home_team'] == team):
                goal_diffs[index] = game['Home_goals'] - game['Away_goals']
            else:
                goal_diffs[index] = game['Away_goals'] - game['Home_goals']
                
        #Calculate the match rating for each match per team
        for index in range(6, len(games)):
            team_ratings[index - 6] = np.sum(goal_diffs[index - 6:index])
            
        #Insert the match rating into the df
        for num in range(6, len(games)):
            game = df.iloc[games[num]]
            if (game['Home_team'] == team):
                df.at[games[num], 'Home_rating'] = team_ratings[num - 6]
            else:
                df.at[games[num], 'Away_rating'] = team_ratings[num - 6]
        
    df["Match_rating"] = df['Home_rating'] - df['Away_rating']
```

### Creating the dataframes

```python
df15_16 = pd.read_csv('PL15-16.csv')
df16_17 = pd.read_csv('PL16-17.csv')
df17_18 = pd.read_csv('PL17-18.csv')
df18_19 = pd.read_csv('PL18-19.csv')

df15_16 = shrink_df(df15_16)
df16_17 = shrink_df(df16_17)
df17_18 = shrink_df(df17_18)
df18_19 = shrink_df(df18_19)
```

### Calculating the match ratings

```python
add_ratings(df15_16)
add_ratings(df16_17)
add_ratings(df17_18)
add_ratings(df18_19)
```

### Creating a new dataframe combining all seasons results and ratings

```python
df1 = df15_16[['Result', 'Match_rating']].dropna()
df2 = df16_17[['Result', 'Match_rating']].dropna()
df3 = df17_18[['Result', 'Match_rating']].dropna()
df4 = df18_19[['Result', 'Match_rating']].dropna()

final_df = pd.concat([df1, df2, df3, df4])
```

### Creating a new dataframe holding totals and %'s for ratings

```python
totals = pd.DataFrame(columns=['Match_rating', 'Wins', 'Draws', 'Losses'])
totals['Match_rating'] = range(-30, 30)
totals['Wins'] = 0
totals['Draws'] = 0
totals['Losses'] = 0

totals = totals.set_index('Match_rating')

for index, row in final_df.iterrows():
    rating = row['Match_rating']
    result = row['Result']
    
    if (result == 'H'):
        totals.at[int(rating), 'Wins'] += 1
    elif (result == 'D'):
        totals.at[int(rating), 'Draws'] += 1
    else:
        totals.at[int(rating), 'Losses'] += 1

totals = totals.reset_index()

totals['Win %'] = round((totals.Wins / (totals.Wins + totals.Draws + totals.Losses)), 2)
totals['Draw %'] = round((totals.Draws / (totals.Wins + totals.Draws + totals.Losses)), 2)
totals['Loss %'] = round((totals.Losses / (totals.Wins + totals.Draws + totals.Losses)), 2)
```

### Visualizing the data

```python
x = totals.dropna()['Match_rating']
a1, b1, r1, p1, std1 = stats.linregress(x, totals.dropna()["Win %"])
a2, b2, c2 = np.polyfit(x, totals.dropna()["Draw %"], deg=2)
a3, b3, c3 = np.polyfit(x, totals.dropna()["Loss %"], deg=2)
```

```python
def odds_calculator(rating):
    w_chance = round(a1 * rating + b1, 2)
    d_chance = round(a2 * rating**2 + b2 * rating + c2, 2)
    l_chance = round(a3 * rating**2 + b3 * rating + c3, 2)

    print("Home win odds:", w_chance)
    print("Draw odds:", d_chance)
    print("Away win odds:", l_chance)
    
def home_odds_calc(rating):
    return round(a1 * rating + b1, 2)

def draw_odds_calc(rating):
    return round(a2 * rating**2 + b2 * rating + c2, 2)

def away_odds_calc(rating):
    return round(a3 * rating**2 + b3 * rating + c3, 2)
```
