```python
df['h_goals'] = df15_16['FTHG']
df['a_goals'] = df15_16['FTAG']
df['h_shots'] = df15_16['HS']
df['a_shots'] = df15_16['AS']
df['h_on_target'] = df15_16['HST']
df['a_on_target'] = df15_16['AST']
df['home_win'] = np.amax(df15_16[['B365H', 'BWH', 'IWH', 'LBH', 'PSH', 'WHH']], 1)
df['draw'] = np.amax(df15_16[['B365D', 'BWD', 'IWD', 'LBD', 'PSD', 'WHD']], 1)
df['away_win'] = np.amax(df15_16[['B365A', 'BWA', 'IWA', 'LBA', 'PSA', 'WHA']], 1)
df['result'] = df15_16['FTR']
```

```python
mydf = pd.DataFrame(columns=['h_goals', 'a_goals', 'h_shots', 'a_shots', 'h_on_target', 'a_on_target', 'home_win', 'draw', 'away_win', 'result'])
for x in range(4):
    temp = pd.read_csv('PL' + str(15 + x) + '-' + str(16 + x) + '.csv')
    temp2 = pd.DataFrame(columns=['h_goals', 'a_goals', 'h_shots', 'a_shots', 'h_on_target', 'a_on_target', 'home_win', 'draw', 'away_win', 'result'])
    temp2['h_goals'] = temp['FTHG']
    temp2['a_goals'] = temp['FTAG']
    temp2['h_shots'] = temp['HS']
    temp2['a_shots'] = temp['AS']
    temp2['h_on_target'] = temp['HST']
    temp2['a_on_target'] = temp['AST']
    temp2['home_win'] = np.amax(temp[['B365H', 'BWH', 'IWH', 'LBH', 'PSH', 'WHH']], 1)
    temp2['draw'] = np.amax(temp[['B365D', 'BWD', 'IWD', 'LBD', 'PSD', 'WHD']], 1)
    temp2['away_win'] = np.amax(temp[['B365A', 'BWA', 'IWA', 'LBA', 'PSA', 'WHA']], 1)
    temp2['result'] = temp['FTR']
    mydf = pd.concat([mydf, temp2])
```
